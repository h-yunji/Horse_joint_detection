```python
def h5_dif(df):    # 현재 열 이름을 새로운 행으로 추가
    df.reset_index(inplace = True)
    df_new = df.T.reset_index().T
    
    # 첫 번째 행을 새로운 열 이름으로 설정
    df_new.columns = df_new.iloc[0]
    
    # 첫 번째 행 제거 (이제 열 이름이 되었으므로)
    df_new = df_new.drop(df_new.index[0])
    
    # 인덱스 재설정 (선택사항)
    df_new = df_new.reset_index(drop=True)
    return df_new
```

```python
import pandas as pd
import numpy as np

def extract_rows_with_nan(csv_file):
    """
    CSV 파일에서 NaN값이 포함된 행만 추출하는 함수
    
    Parameters:
    csv_file (str): CSV 파일 경로
    
    Returns:
    DataFrame: NaN값이 포함된 행들만 있는 DataFrame
    """
    # CSV 파일 읽기
    df = csv_file
    
    # NaN이 포함된 행만 필터링
    rows_with_nan = df[df.isna().any(axis=1)]
    
    # NaN이 있는 열 찾기
    for index, row in rows_with_nan.iterrows():
        nan_columns = row.index[row.isna()].tolist()
        print(f"\n행 {index}의 NaN이 있는 열:", nan_columns)
    
    return rows_with_nan

# 사용 예시
if __name__ == "__main__":
    # NaN이 있는 행 추출
    result = extract_rows_with_nan(h_csv7)
    print("\n추출된 데이터:")
    print(result)
```

```python
import json
import glob
import os

save_path = "/RnD/horse/" # COCO dataset을 저장할 경로
save_as = 'horse_trains.json'
output_path = os.path.join(save_path, save_as)
img_path_sepa="/RnD/horse/"

# COCO 형식 정의
coco_format = {
    "images": [],
    "annotations": [],
    "categories": [
        {
            "id": 1,
            "name": "horse",
            "supercategory": "animal",
            "keypoints": ["Nostril",
                            "Eye",
                            "Poll",
                            'Withers',
                            'LowestBack',
                            'T16L1'
                             ,'T_sacrale'
                             ,'Tail_Root'
                             ,'T_ischiadicum'
                             ,'Tub'
                             ,'Spina_scapulae'
                             ,'ElbowJoint_L'
                             ,'ElbowJoint_R'
                             ,'Carpuse_L'
                             ,'Carpuse_R'
                             ,'Fetlock_L'
                             ,'Fetlock_R'
                             ,'Front_Heel_L'
                             ,'Front_Heel_R'
                             ,'Front_Toe_L'
                             ,'Front_Toe_R'
                             ,'Abdomen'
                             ,'T_Coxae'
                             ,'Coxofemoral'
                             ,'Stifle_Joint_L'
                             ,'Stifle_Joint_R'
                             ,'Rear_Tarsus_L'
                             ,'Rear_Tarsus_R'
                             ,'Rear_Fetlock_L'
                             ,'Rear_Fetlock_R'
                             ,'Rear_Heel_L'
                             ,'Rear_Heel_R'
                             ,'Rear_Toe_L'
                             ,"Rear_Toe_R"],
            "skeleton": []
        }
    ]
}

lst = [h_csv, h_csv2, h_csv5, h_csv6, h_csv7, h_csv8, h_csv11, h_csv4, h_csv9, h_csv10]
# lst = [h_csv3, h_csv12]
annotation_id = 0
for csv in lst :    
    for i, image_name in enumerate(csv.loc[2:,'image_name']):
        annotation_id += 1
    
        # 이미지 정보 추가
        image_info = {
            "id": annotation_id,
            "width": 1920,
            "height": 1080,
            "file_name": os.path.join(img_path_sepa,image_name)
        }
        coco_format["images"].append(image_info)
    
        # 어노테이션 정보 추가
        keypoints = []  # 17*3 (x,y,group_id)
        bbox = []
        
        for box in csv.iloc[2+i, 71:75]:
            bbox.append(box)
        
        for j, label in enumerate(csv.iloc[2+i, 3:71]):
            keypoints.append(float(label))
            if (j + 1) % 2 == 0:  # 3의 배수인 인덱스(0-based) 다음에 2를 추가
                keypoints.append(2)
    
    
        annotation = {
            "id": annotation_id,
            "image_id": annotation_id,
            "category_id": 1,
            "area": bbox[2] * bbox[3],
            "bbox": bbox,
            "iscrowd": 0,
            "keypoints": keypoints,
            "num_keypoints": 34
        }
    
        coco_format["annotations"].append(annotation)


# 변환된 데이터 저장
with open(output_path, 'w') as f:
    json.dump(coco_format, f, indent=4)

print(f"COCO dataset saved to {output_path}")
```

```python
def find_json_error(file_path, target_line=None):
    try:
        # 파일을 줄 단위로 읽기
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()

        # 특정 줄만 검사하려는 경우
        if target_line:
            if 0 <= target_line < len(lines):
                print(f"\n=== 검사 중인 줄 {target_line + 1} ===")
                print(f"내용: {lines[target_line].rstrip()}")
                # 해당 줄의 각 문자 위치 출력
                for i, char in enumerate(lines[target_line]):
                    print(f"위치 {i+1}: '{char}' (ASCII: {ord(char)})")
            return

        # 전체 파일 검사
        print("=== 전체 파일 JSON 유효성 검사 중 ===")
        
        # 줄 단위로 검사하면서 문제가 있는 부분 찾기
        accumulated_data = ""
        for i, line in enumerate(lines):
            accumulated_data += line
            try:
                # 현재까지의 데이터가 유효한 JSON인지 테스트
                json.loads(accumulated_data)
            except json.JSONDecodeError as e:
                print(f"\n문제 발견 - 줄 {i+1}:")
                print(f"줄 내용: {line.rstrip()}")
                
                # 해당 줄의 문자들을 하나씩 검사
                suspicious_chars = []
                for col, char in enumerate(line):
                    if ord(char) not in [9, 10, 13, 32] and not (32 <= ord(char) <= 126):
                        suspicious_chars.append((col+1, char, ord(char)))
                
                if suspicious_chars:
                    print("\n의심스러운 문자들:")
                    for col, char, ascii_val in suspicious_chars:
                        print(f"열 {col}: '{char}' (ASCII: {ascii_val})")
                
                # 주변 컨텍스트 출력
                context_start = max(0, i-2)
                context_end = min(len(lines), i+3)
                print("\n주변 컨텍스트:")
                for ctx_i in range(context_start, context_end):
                    prefix = ">> " if ctx_i == i else "   "
                    print(f"{prefix}줄 {ctx_i+1}: {lines[ctx_i].rstrip()}")
                
                return

    except FileNotFoundError:
        print(f"파일을 찾을 수 없습니다: {file_path}")
    except Exception as e:
        print(f"예상치 못한 에러 발생: {str(e)}")
```
